```python
#Splitting the dataset in training and testing sets
y = df_raw['Class']
X = df_raw.drop('Class', axis = 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2,random_state = 0)
print(len(X_train), len(X_test))
print(len(y_train), len(y_test))
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
#Fitting the training set in Random Forest Classifier model

rfc = RandomForestClassifier(random_state=1000)
rfc.fit(X_train, y_train)
predictions = rfc.predict(X_test)
print("Accuracy: ", accuracy_score(y_test, predictions))
print(classification_report(y_test, predictions))
print(confusion_matrix(y_test, predictions))
```

```python
# saving the results to 'metrics.txt' file

with open('metrics.txt', 'wb') as c:
    pickle.dump(rfc, c, pickle.HIGHEST_PROTOCOL)
```
